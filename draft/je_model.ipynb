```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix, f1_score, roc_auc_score

# 혼동행렬, 정확도, 정밀도, 재현율, F1, AUC 불러오기
def get_clf_eval(y_test, y_pred, pred_proba):
    confusion = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    F1 = f1_score(y_test, y_pred)
    AUC = roc_auc_score(y_test, pred_proba)
    print('오차행렬:\n', confusion)
    print('\n정확도: {:.4f}'.format(accuracy))
    print('정밀도: {:.4f}'.format(precision))
    print('재현율: {:.4f}'.format(recall))
    print('F1: {:.4f}'.format(F1))
    print('AUC: {:.4f}'.format(AUC))
```

```python
# 데이터 나누기 - 6:2:2 비율
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1,stratify = y)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1, stratify = y_train)
print(X_train.shape, X_val.shape, X_test.shape)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=0)
X_train_over,y_train_over = smote.fit_sample(X_train,y_train)
print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트: ', X_train.shape, y_train.shape)
print('SMOTE 적용 후 학습용 피처/레이블 데이터 세트: ', X_train_over.shape, y_train_over.shape)
print('SMOTE 적용 후 레이블 값 분포: \n', pd.Series(y_train_over).value_counts())
print(X_train_over.shape, X_val.shape, X_test.shape)
```

## SVM

```python
# 여러 가지 kernel을 사용해 SVM 학습하고 accuracy 계산하기
kernels = ['poly','rbf','sigmoid']

for kernel in kernels:
    print("==========================================") 
    model = SVC(kernel = kernel)
    model.fit(X_train_over, y_train_over)
    pred_val = model.predict(X_val)
    print(f"accuracy of {kernel}: {accuracy_score(y_val, pred_val)}")
```

```python
# 여러 가지 kernel을 사용해 SVM 학습하고 accuracy 계산하기
model = SVC(kernel = 'rbf', C=10000, gamma=0.01) # 파라이터값 그리드서치한 결과임 
model.fit(X_train_over, y_train_over)
new_pred_test = model.predict(X_test)
print(f'accuracy: {accuracy_score(y_test, new_pred_test)}')
```

```python
def LIFT(preds, y_test): # >=2.5
    condition = y_test.astype(bool) #preds에 정답인 예측 score만 남긴다.
    c = np.extract(condition, preds)
    b = np.argsort(-c)[:len(c)//5] #예측 score 상위 20%
    
    lift_20 = preds[b] 
    
    #lift_20_flat = np.argmax(lift_20, axis=1) #값을 확률에서 0, 1값으로 바꾼다. 
    #y_test_flat = np.argmax(y_test, axis=1)
    #print(lift_20_flat)
    lift_20_1 = lift_20[lift_20 == 1] #1인 경우만 남긴다. 
    y_1 = y_test[y_test == 1]
    
    print('LIFT Accuracy: ',  (len(lift_20_1)/len(lift_20))/(len(y_1)/len(y_test)))
```
